- `retrain.py`가 존재하는 디렉토리에서 cmd 를 실행시키고 다음 명령을 입력한다.

``` sh
python retrain.py \ --bottleneck_dir=./workspace/bottlenecks \ --model_dir=./workspace/inception \ --output_graph=./workspace/flowers_graph.pb \ --output_labels=./workspace/flowers_labels.txt \ --image_dir ./workspace/flower_photos \ --how_many_training_steps 1000
```

- 스크립트 옵션 설명
    - `--bottleneck_dir` : 학습할 사진을 인셉션용 학습 데이터로 변환해서 저장해둘 디렉터리
    - `--model_dir` : 인셉션 모델을 내려받을 경로
    - `--output_graph` : 학습된 모델을 저장할 경로
    - `--output_labels` : 레이블 이름들을 저장해둘 파일 경로
    - `--image_dir` : 원보 이미지 경로
    - `--how_many_training_steps` : 반복 학습 횟수

- 학습이 상당히 오래 진행된다.

![image](https://user-images.githubusercontent.com/43382559/75746156-b992c700-5d5c-11ea-957c-fb7a66fe6980.png)

---

### 실패 & 해결

- 다음과 같은 에러가 뜬다.

![image](https://user-images.githubusercontent.com/43382559/75747440-48551300-5d60-11ea-9194-63ffcec60fd5.png)



- [tensorflow-hub_retrain.py](https://github.com/tensorflow/hub/blob/master/examples/image_retraining/retrain.py)
최신스크립트로 다시 시도해보자

- contrib 모듈을 찾을수 없다는 에러는 `pip install tensorflow==1.14` 명령으로 해결했다.

### 결과

![image](https://user-images.githubusercontent.com/43382559/75750178-eea41700-5d66-11ea-8580-569685e78fa4.png)

## 11.3 예측스크립트 실행

- `python predict.py (파일경로)` 로 실행하면 된다고 하는데

![image](https://user-images.githubusercontent.com/43382559/75751182-39269300-5d69-11ea-9e63-96b7a45d6bce.png)

-이런 에러가 뜬다.

- 찾아보니 이런 문제가 있어 feed_dict 을 실행할수 없다고 한다.[stackoverflow](https://stackoverflow.com/questions/45660439/the-name-decodejpeg-contents0-refers-to-a-tensor-which-does-not-exist)

- 해결방법을 더 찾아봐야겠다.
